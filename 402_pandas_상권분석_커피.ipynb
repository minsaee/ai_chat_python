```python
# 이디야와 할리스 데이터 가져오기
df_coffee_edi_hollys = df_coffee[df_coffee['상호명'].str.contains('할리스|이디야')]
df_coffee_edi_hollys.shape
```

```python
df_coffee_edi_hollys.sample(5)
```

```python
# 브랜드명 column 추가하기.
df_coffee_edi_hollys.loc[df_coffee_edi_hollys['상호명'].str.contains('이디야'),'브랜드명'] = '이디야'
```

```python
df_coffee_edi_hollys.loc[df_coffee_edi_hollys['상호명'].str.contains('할리스'),'브랜드명'] = '할리스'
```

```python
df_coffee_edi_hollys.sample(5)
```

```python
df_coffee_edi_hollys['브랜드명'].unique()
```

```python
df_coffee_edi_hollys['브랜드명'].value_counts()
```

```python
geo_df_coffee = df_coffee_edi_hollys.copy()
```

```python
geo_df_coffee.sample(5)
```

```python
geo_df_coffee['브랜드명'].value_counts()
```

```python
import folium
```

```python
map = folium.Map(location=[geo_df_coffee['위도'].mean(),geo_df_coffee['경도'].mean()],
                 zoom_start=12)

for n in geo_df_coffee.index:
  # 팝업창에 들어갈 텍스트 지정
  popup_name=geo_df_coffee.loc[n, '상호명'] + ' - '\
  + geo_df_coffee.loc[n, '도로명']

  if geo_df_coffee.loc[n, '브랜드명']=='이디야':
    icon_color = 'blue'
  else:
    icon_color = 'red'

  folium.CircleMarker(
      location=[geo_df_coffee.loc[n, '위도'],geo_df_coffee.loc[n, '경도']],
      radius = 3,
      popup = popup_name,
      color = icon_color,
      fill = True,
      fill_color = icon_color
  ).add_to(map)

# 배경지도 타일 설정하기
tiles = "CartoDB dark_matter"
# 배경지도 타일 레이어를 지도에 추가하기
folium.TileLayer(tiles=tiles).add_to(map)

map
```
