### 변수
기계학습에서는  column = feature = dimension = variable을 혼용해서 사용한다.  


1 범주형 변수(Categorical Data)=질적변수(Qualitative Variable) : 몇개의 범주로 나누어진 자료를 의미
 - 명목형(Nominal Data): 성별, 성공여부, 혈액형 등 순서를 매길 수 없고 셀수 있다는 특징이 있다.
 - 순서형(Ordinal Data) : 교육수준, 만족도등  그들 사이에 순서 관계가 존재하는 자료


2 수치형 변수 = 양적변수(Quantitative Variable): 이산형과 연속형으로 이루어진 자료를 의미
  - 이산형(Discrete Data) :출산횟수, 불량품 수, 사고건수, 차량대수 등 이산적인 값을 갖는 데이터로  수치적인 의미를 가지나 소수점의 형태로 표현되지 못하는 데이터를 의미
  - 연속형(Continuous Data) : 신장, 체중, 시간, 길이 등 연속적인 값을 가지고 소수점으로 표현이 가능한 데이터를 의미




|        |  질적변수(Category)      | 양적변수(Numeric)          |
|--- |--- |---|
|질적변수|Cross Table               | Box Plot                   |
|        |Mosaic Plot               | 범주(Category)별 통계분석  |
|양적변수|Box Plot                  | Correlation                |
|        |범주(Category)별 통계분석 | Regression                 |
|        |                          | Scatter Plot               |


더미변수란 범주형 변수에 있는 범주 각각을 컬럼으로 변경하고, 원본 컬럼의 값이 해당 범주에 속하는지 여부에 따라 1 혹은 0으로 채운 변수이다.  




pd.get_dummies(df, columns=['범주형변수'])
- 범주형 변수 변환
  - 원핫인코딩(One-hot-encoding), 더미코딩(dummy coding),   이펙트코딩(Effect coding), 숫자로 표현된 범주형 특성,   레이블인코딩(Label encoding), 특징 해싱(Feature Hashing)  


- 수치형 변수 변환 (*옆의 괄호는 Scikit Learn의 함수명)
  - 선형변환: 최소최대 스케일링(MinMaxScaler), 표준화(StandardScaler), 로버스트 스케일링(RobustScaler), 균등분포/RankGauss (QuantileTransformer)
  - 비선형변환: 로그변환, 거듭제곱변환(PowerTransformer - Boxcox, YeoJohnson), 정규화(Normalizer - L1, L2, Max)
  - 기타: 구간분할 (=이산화, binning), 순위 변환



```python
# iris데이터의 target값인 범주형 변수를 더미변수 형태로 대체하기.

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris

load_iris = load_iris()
# print(load_iris)
print(dir(load_iris))
print(type(load_iris))
# print(load_iris.DESCR)
```

```python
iris = pd.DataFrame(load_iris.data, columns = load_iris.feature_names)
iris['target'] = load_iris.target
iris.sample(5)
```

```python
iris.info()
```

```python
iris['target'] = iris['target'].replace([0,1,2],['Setosa','Versicolour','Virginica'])
# iris['target'] = iris['target'].map({0:'Setosa', 1:'Versicolour', 2:'Virginica'})
```

```python
iris['target']
```

```python
iris.sample(5)
```

```python
# iris dataset의 target변수의 값을 더미로 처리하기.
iris_dummy = pd.get_dummies(iris, columns = ['target'])
iris_dummy.sample(5)
```

### label encoding
 - label encoding은 문자열 데이터에 고유한라벨 번호를 매기는 것이다.

```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.sample(5)
```

```python
# 실제 데이터 값이 바뀌지는 않는다.
iris.species.map({'versicolor':0, 'virginica':1, 'setosa':2})
```

```python
iris.sample(5)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
ll = LabelEncoder()
```

```python
# ll.fit_transform(iris.species)
ll.fit_transform(iris['species'])
```

```python
# fancy indexing을 이용한 2차원으로 입력하면 경고창이 뜬다.
# A column-vector y was passed when a 1d array was expected. Please change the shape of y to (n_samples, ), for example using ravel(). y = column_or_1d(y, warn=True)
# ll.fit_transform(iris[['species']])
```

```python
from sklearn.utils import column_or_1d

column_or_1d(iris['species'])
ll.fit_transform(column_or_1d(iris['species']))
```

### one hot encoding
 - 문자열 데이터를 column으로 바꾸고, 해당하는 위치에만 1로한다.

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

sparse matrix
 - 성긴 행렬은 행렬의 값이 대부분 0인 경우를 가리키는 표현이다.
 - 그와 반대되는 표현으로는 밀집행렬, 조밀행렬이 사용된다.
 - 개념적으로 성김은 시스템들이 약하게 연결된 것에 해당한다.
 -  한 줄로 나열된 공과 공이 스프링으로 양 옆으로 하나씩 연결되었을 때 이것은 성긴 시스템이다.

```python
# 2차원
ohe.fit_transform(iris[['species']])
```

```python
ohe.fit_transform(iris[['species']]).toarray()
```

```python
# inverse_tranform으로 복구 가능하다.
ohe.inverse_transform([[1,0,0],[0,1,0],[0,0,1]])
```
