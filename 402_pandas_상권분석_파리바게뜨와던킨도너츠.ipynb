```python
# 필요한 columns만 세팅하기.
dataset = df[['상호명','지점명','상권업종중분류명','시도명','시군구명','행정동명','도로명','경도','위도']]
dataset.head()
```

```python
# 상호명만 가져오기.
dataset['상호명'].value_counts()
```

```python
# 상호명에서 '파리바게뜨'가 있는것만 가져오기.
# 상권업종중분류명은 '기타간이' 이다.
dataset[dataset['상호명'].str.contains('파리바게뜨')]
```

```python
# 상호명에서 '던킨도너츠'가 있는것만 가져오기.
# 상권업종중분류명이 기타 간이 이다.
dataset[dataset['상호명'].str.contains('던킨도너츠')]
```

```python
# 파리바게뜨와 던킨도너츠 데이터 가져오기.
paris_dunkin = dataset[dataset['상호명'].str.contains('파리바게뜨|던킨도너츠')]
paris_dunkin.shape
```

```python
paris_dunkin.sample(5)
```

```python
# column에 브랜드명 추가하기.
paris_dunkin.loc[paris_dunkin['상호명'].str.contains('파리바게뜨'),'브랜드명'] = '파리바게뜨'
```

```python
paris_dunkin.loc[paris_dunkin['상호명'].str.contains('던킨도너츠'),'브랜드명'] = '던킨도너츠'
```

```python
paris_dunkin.sample(5)
```

```python
paris_dunkin['브랜드명'].value_counts()
```

```python
import folium
```

```python
map = folium.Map(location=[paris_dunkin['위도'].mean(),paris_dunkin['경도'].mean()],
                 zoom_start=12)

for n in paris_dunkin.index:
  popup_name=paris_dunkin.loc[n,'상호명']+ '-'\
  +paris_dunkin.loc[n,'도로명']

  if paris_dunkin.loc[n,'브랜드명']=='파리바게뜨':
    icon_color = 'blue'
  else:
    icon_color = 'pink'

  folium.CircleMarker(
      location=[paris_dunkin.loc[n, '위도'], paris_dunkin.loc[n, '경도']],
      radius = 3,
      popup = popup_name,
      color = icon_color,
      fill = True,
      fill_color = icon_color
  ).add_to(map)

# 배경지도 타일 설정하기
tiles = "CartoDB dark_matter"
# 배경지도 타일 레이어를 지도에 추가하기
folium.TileLayer(tiles=tiles).add_to(map)

map
```
