## at
 - 이름으로 한 위치의 값만 가져온다.
 - at은 slicing을 사용할 수 없다.
 - at은 fancy indexing을 사용할 수 없다.

```python
# print(data['tip',0]) # KeyError: ('tip', 0)
print(data.at[0,'tip'])
print(data.loc[0,'tip'])
print(data.iloc[0,1])


print(data['tip'][0])
# print(data.at[0]['tip']) # TypeError: DataFrame._get_value() missing 1 required positional argument: 'col'
print(data.loc[0]['tip'])
print(data.iloc[0][1])

print(data['tip']) # 열(이름)
print(data[3:5]) # 행(번호)
```

```python
# column

print(data[['tip','day']])
# data['tip':'day'] # TypeError

# rows
print(data[0:5])
# data[[0,1]] # KeyError
```

where

```python
a = np.array([10,20,30,40,50])
a
```

```python
# where의 인자는 조건이다.
# 조건만 있을 경우 True의 위치를 return

np.where(a>30)
```

```python
# where의 인자 개수에 따라서 기능이 달라진다.
# 조건이 True이면 a+1, False이면 a를 return

np.where(a>30, a+1, a)
```
