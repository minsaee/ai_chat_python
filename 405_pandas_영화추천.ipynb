```python
# 본 영화목록 함수로 정의하기.
def movie_seen(user_id):
  return recommendation_pivot.loc[user_id][recommendation_pivot.loc[user_id]>0]
```

corr (correlationship)
 - corr은 column기준으로 상관관계를 분석해준다.

```python
users[['UserID','Age']].corr()
```

```python
# UserID를 기준으로 상관관계를 분석을 하기 위해 행과 열을 바꿔준다.
small_test = recommendation_pivot.T.iloc[:500,:500]
small_test
```

```python
small_test_corr = small_test.corr()
small_test_corr
```

```python
# UserID 5와 상관관계가 높은 상위 10개 출력하기.
# sort_values 정렬
# ascending의 default는 True(오름차순)이다. False(내림차순)
# 슬라이싱에서 index 0은 제외하고 출력한다. -> index 0은 자기자신
small_test_corr[5].sort_values(ascending=False)[1:11]
```

```python
# 함수 정의를 통한 일반화
def nearest_user(user_id, n):
  return small_test_corr.loc[user_id].sort_values(ascending = False)[1:n+1]
```

```python
# 5번 ID와 상관관계가 높은 3명 출력하기.
nearest_user(5,3)
```

```python
# 상관관계가 높은 193번과 5번의 유저가 본 영화 비교하기.
# 193번 유저가 본 영화
print(movie_seen(193))
# 5번 유저가 본 영화
print(movie_seen(5))
```

```python
# 193번은 보고 5번은 보지 않은 영화목록 출력하기.
unseen = set(movie_seen(193).index)-set(movie_seen(5).index)
len(unseen)
```

```python
unseen_movie = recommendation_data[recommendation_data.MovieID.isin(unseen)]
unseen_movie
```

```python
# 193번의 Rating 5를 merge하기.
unseen_movie[(unseen_movie.Rating==5)&(unseen_movie.UserID==193)].merge(movies)
```

```python
# 상관관계가 있는 유저끼리 상대방이 보지않은 영화 중 평점이 5점인 영화를 추천하기.
# 함수로 정의하기.

def unseen_movie(user_id, other_id):
  unseen = set(movie_seen(user_id).index)-set(movie_seen(other_id).index)
  unseen_movie = recommendation_data[recommendation_data.MovieID.isin(unseen)]
  return unseen_movie[(unseen_movie.Rating==5)&(unseen_movie.UserID==user_id)].merge(movies)
```

```python
# 함수로 정의하기.

def recommend_movie(user_id, n):
  other_list = nearest_user(user_id, n).index
  other_mv_list = recommendation_data[(recommendation_data.UserID.isin(other_list))&(recommendation_data.Rating==5)]
  print(other_mv_list.columns)

  user_mv_list = movie_seen(user_id)

  unseen_list = set(other_mv_list['MovieID'])-set(user_mv_list.index)
  return movies[movies['MovieID'].isin(unseen_list)].reset_index(drop=True)
```
